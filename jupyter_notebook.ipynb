## Raças de Gato Mais Caras

Plot Top 10 raças mais caras

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados da tabela
data = {
    'Nome da raça': [
        'Extra-Toes Cat (Hemingway Polydactyl)', 'Exotic Shorthair', 'British Shorthair', 
        'Maine Coon', 'Applehead Siamese', 'Chausie', 'Munchkin', 
        'Canadian Hairless', 'Sphynx (hairless cat)', 'American Curl'
    ],
    'Média de preço': [2000, 1750, 1750, 1600, 1600, 1500, 1500, 1450, 1450, 1400]
}

# Criar um DataFrame
df = pd.DataFrame(data)

# Configurar o tamanho do gráfico
plt.figure(figsize=(10,6))

# Criar um gráfico de barras
plt.barh(df['Nome da raça'], df['Média de preço'], color='#FFA07A')

# Adicionar título e rótulos
plt.title('Raças de Gatos Mais Caras', fontsize=14)
plt.xlabel('Média de Preço ($)', fontsize=12)
plt.ylabel('Nome da Raça', fontsize=12)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

## Agrupamento Pelo Peso

Preço médio

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados da tabela
data = {
    'Peso (kg)': [7.00, 6.00, 5.00, 4.00, 3.00],
    'Preço Médio ($)': [1337.50, 612.96, 1100.00, 928.13, 1112.50]
}

# Criar um DataFrame
df = pd.DataFrame(data)

# Configurar o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Criar um gráfico de dispersão
plt.scatter(df['Peso (kg)'], df['Preço Médio ($)'], color='orange', s=100)

# Adicionar título e rótulos
plt.title('Preço Médio em Função do Peso', fontsize=14)
plt.xlabel('Peso (kg)', fontsize=12)
plt.ylabel('Preço Médio ($)', fontsize=12)

# Adicionar grid
plt.grid()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
